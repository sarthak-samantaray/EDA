```python
missing_value_per = df.isnull().sum() / df.shape[0] *100
missing_value_per
```

```python
# dropping columns 
drop_colums = missing_value_per[missing_value_per > 20].keys()
drop_colums
```

```python
df2_drop_columns = df.drop(columns = drop_colums)
```

```python
df2_drop_columns.shape
```

```python
# accesing numeric variable
df3_num = df2_drop_columns.select_dtypes(include=['int64','float64'])
df3_num.head()
```

```python
plt.figure(figsize = (16,9))
sns.heatmap(df3_num.isnull())
```

```python
df3_num[df3_num.isnull().any(axis = 1)]
df3_num
```

```python
df3_num.isnull().sum()
```

```python
missing_num_var = []
for var in df3_num.columns:
    if df3_num[var].isnull().sum() > 0:
        missing_num_var.append(var)
missing_num_var
```

```python
# to find which way either median or mean to use. 
# we first go for the distribution of the data.

plt.figure(figsize = (10,10))
#sns.set()
for i ,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins = 20,kde_kws={'linewidth':5, 'color': '#DC143C'})
```

```python
df4_num_mean = df3_num.fillna(df3_num.mean())
```

```python
df5_num_median = df3_num.fillna(df3_num.median())
```
