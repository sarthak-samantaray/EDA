```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:/Datasets/titanic.csv",usecols = ['Cabin','Survived'])
```

```python
df.head()
```

# Replace Nan value with "Missing".

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df
```

```python
df['Cabin'] = df['Cabin'].astype(str).str[0]
```

```python
df
```

##### Performing Probability Ratio Encoding

```python
###Finding the percentage of the survived people according to Cabin
df.groupby(['Cabin'])['Survived'].mean()

### this can be called as probablity.
```

```python
prob_df = df.groupby(['Cabin'])['Survived'].mean()
prob_df
```

```python
prob_df = pd.DataFrame(prob_df,columns=['Survived'])
```

```python
prob_df
```

```python
### To calculate , waht is the probaility that the person has died.
prob_df['Died'] = 1 - prob_df['Survived']
prob_df
```

```python
prob_df['Probability_Ratio'] = prob_df['Survived'] / prob_df['Died']
```

```python
prob_df
```

```python
probablity_encoed = prob_df['Probability_Ratio'].to_dict()
probablity_encoed
```

```python
df['Cabin_encoded'] = df['Cabin'].map(probablity_encoed)
```
