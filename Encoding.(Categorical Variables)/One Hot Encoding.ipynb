We can see that from just 6 initital categorical variables, we end up with 117 new variables.

What can we do??

We can find out 10 most frequent categories in each feature and use that, and skip the other categories in each feature.<br>
For all the remaining categories it will be zero.

```python
## Let's find the top 10 most frequent categories for the variable x2
df.X2.value_counts().sort_values(ascending=False).head(10)
```

```python
## Let's make a list with the most frequent categories of the variable.

top_10 = [x for x in df.X2.value_counts().sort_values(ascending=False).head(10).index]
```

```python
top_10
```

```python
## Now we make 10 binary variables.

for label in top_10:
    df[label] = np.where(df['X2']==label,1,0)
    
df[['X2']+top_10].head(40)
```

```python
### get whole set of dummy variables for all categorical variables.

def one_hot_top_x(df,variable,top_x_labels):
    ## Function to create dummy variables for the most frequent labels.
    ## We can vary the number of most frequent labels that we encode.
    
    for label in top_x_labels:
        df[variable+'_'+label] = np.where(df[variable]==label,1,0)
        
# read the data again.
df = pd.read_csv("D:/Datasets/mercedes.csv",usecols=['X1','X2','X3','X4','X5','X6'])

# find the top_10 labels manually and then pass it to the above function.
top_10 = [x for x in df['X2'].value_counts().sort_values(ascending=False).head(10).index]

# encode X2 into the 10 most frequent categories.
one_hot_top_x(df,'X2',top_10)
```

```python
top_10_X1 = [x for x in df['X1'].value_counts().sort_values(ascending=False).head().index]
```

```python
one_hot_top_x(df,'X1',top_10_X1)
```

# Advantages.

1. Straightforward to implement
2. Does not reuire hrs of variable exploration
3. Does not expand maassively the feature space(number of columns in the dataset)

# Disadvantages.

1. Does not add any information that make the variable more predictive.
2. Does not keep the information of the ignored labels.

Because it is not unusual that categorical variables have a few domaination categories and the remaining labels add mostly
noise, this is a qiote simple and straighforward approach that may be useful on many occassions. <br>
<br>
It is worth nothing that the top 10 variables is totally arbitary number. You could also choose top 5 or top 20.
