# Mixed Variables
1. Suppose there is a column with a data of :<br>
    A67,C87,A109A3
    How to handle these data?

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("E://Datasets/titanic.csv",usecols=['Cabin','Ticket','Survived'])
```

```python
df = df.sample(10)
```

```python
df['number'] = ['A',1,2,4,8,1,'A',1,2,"A"]
```

In number column there are, numerical data as well as categorical data. 
we need to handle that first.

```python
# Extract the numerical data from a COLUMN.
df['number_numerical'] = pd.to_numeric(df['number'],errors='coerce',downcast='integer')
df
```

```python
# Extracting Categorical Data.
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df
```

Cabin column has data like b80,c98 like this we need seperate this.

```python
# Extracting numerical data from a string.
df['cabin_num'] = df["Cabin"].str.extract('(\d+)')
# capture the first letter.
df['cabin_cat'] = df['Cabin'].str[0]
```

```python
df
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
